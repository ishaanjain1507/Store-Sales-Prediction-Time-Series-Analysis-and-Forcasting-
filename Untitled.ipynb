```python
oil = pd.read_csv("./oil.csv")
oil
```

```python
def plot_df(df, x, y, title="", xlabel='Date', ylabel='OIL', dpi=100):
    plt.figure(figsize=(15,4), dpi=dpi)
    plt.plot(x, y, color='tab:red')
    plt.gca().set(title=title, xlabel=xlabel, ylabel=ylabel)
    x_ticks = [df["date"].iloc[30*i] for i in range(int(df["date"].shape[0]/30)+1)]
    plt.xticks(x_ticks)
#     plt.show()

plot_df(oil, oil["date"], oil["dcoilwtico"])
plt.axhline(oil["dcoilwtico"][:5].mean())
plt.axhline(oil["dcoilwtico"][-6:].mean())
plt.xticks(rotation = 75)
plt.show()
sns.displot(oil, kde = True)
```

```python
training_data = pd.read_csv("./train.csv")
training_data
```

```python
training_data['date'] = pd.to_datetime(training_data['date'])
training_data['year'] = training_data['date'].dt.year
training_data['month'] = training_data['date'].dt.month
training_data['week'] = training_data['date'].dt.isocalendar().week
training_data['quarter'] = training_data['date'].dt.quarter
training_data['day_of_week'] = training_data['date'].dt.day_name()
```
